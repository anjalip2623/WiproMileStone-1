```
def ispalindrome(name):
    text = name.replace(" ", "")
    return text == text[::-1]

def count_the_vowels(name):
    vowels = "aeiouAEIOU"
    count = 0
    for char in name:
        if char in vowels:
            count += 1
    return count

def frequency_of_letters(name):
    freq = {}
    for char in name:
        if char != " ":
            if char in freq:
                freq[char] += 1
            else:
                freq[char] = 1

    result = []
    for k, v in freq.items():
        result.append(f"{k}-{v}")

    return ", ".join(result)
```

```
name_input = input("Enter name: ")

if ispalindrome(name_input):
    print("Yes it is a palindrome.")
else:
    print("No it is not a palindrome.")

vowel_count = count_the_vowels(name_input)
print(f"No of vowels: {vowel_count}")

letter_freq = frequency_of_letters(name_input)
print(f"Frequency of letters: {letter_freq}")
```

```
import sys

sys.argv = ['solution.py', '3-1', '5-7', '1-5-3-8']

def calculate_happiness():
    like_nums = set(sys.argv[1].split('-'))
    dislike_nums = set(sys.argv[2].split('-'))
    given_nums = sys.argv[3].split('-')

    happiness = 0

    for num in given_nums:
        if num in like_nums:
            happiness += 1
        elif num in dislike_nums:
            happiness -= 1

    print(happiness)

if __name__ == "__main__":
    calculate_happiness()
```

```
import collections
import string

def parse_meeting_details(filename="Sample.txt"):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            lines = file.readlines()
    except FileNotFoundError:
        print(f"Error: {filename} not found.")
        return
    line_count = len(lines)

    if line_count == 0:
        print("The file is empty.")
        return
    hour_24 = line_count % 24
    if hour_24 == 0:
        time_str = "12 AM"
    elif hour_24 == 12:
        time_str = "12 PM"
    elif hour_24 > 12:
        time_str = f"{hour_24 - 12} PM"
    else:
        time_str = f"{hour_24} AM"
    word_counts = collections.Counter()

    for line in lines:
        cleaned_line = line.translate(str.maketrans('', '', string.punctuation))
        words = cleaned_line.split()

        for word in words:
            if word:
                word_counts[word] += 1

    if not word_counts:
        print("No words found in the file.")
        return

    most_common_word, max_count = word_counts.most_common(1)[0]
    place_name = most_common_word.capitalize()
    print(f"Meeting time: {time_str}")
    print(f"Meeting place: {place_name} Street")
    print(f"\nExplanation: Number of lines is {line_count}. "
          f"The word {most_common_word} appears for the maximum of {max_count} times.")

if __name__ == "__main__":
    parse_meeting_details("Sample.txt")
```

```
def read_n_lines(filename="sample.txt"):
    try:
        n = int(input("Enter the number of lines to read: "))

        with open(filename, 'r', encoding='utf-8') as file:
            print(f"--- First {n} Lines ---")
            for i in range(n):
                line = file.readline()
                if not line:
                    break  # Reached End of File (EOF)
                print(line, end='')  # end='' prevents double newlines
    except ValueError:
        print("Please enter a valid integer.")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")

if __name__ == "__main__":
    read_n_lines()
```

5. Find longest word

```
import string

def find_longest_word(filename="sample.txt"):
    try:
        with open(filename, 'r', encoding='utf-8') as file:
            content = file.read()
        cleaned_content = content.translate(str.maketrans('', '', string.punctuation))
        words = cleaned_content.split()
        if not words:
            print("The file contains no words.")
            return
        longest_word = max(words, key=len)
        print(f"The longest word is: '{longest_word}' (Length: {len(longest_word)})")
    except FileNotFoundError:
        print(f"Error: The file '{filename}' does not exist.")
if __name__ == "__main__":
    find_longest_word()
```

```
try:
    filename = input("Enter the file name (without .txt): ").strip()
    if not filename.endswith(".txt"):
        filename += ".txt"
    total_items = 0
    free_items = 0
    amount = 0
    discount = 0
    with open(filename, "r") as file:

        for line in file:
            line = line.strip()
            if line == "":
                continue
            parts = line.split()

            if len(parts) != 2:
                continue
            item = parts[0]
            price = parts[1]
            if item.lower() == "discount":
                try:
                    discount = float(price)
                except ValueError:
                    print("Invalid discount value.")
                continue
            if price.lower() == "free":
                free_items += 1
                continue
            try:
                amount += float(price)
                total_items += 1
            except ValueError:
                print(f"Invalid price for {item}")

    final_amount = amount - discount
    print("\n------ Purchase Summary ------")
    print("No of items purchased :", total_items)
    print("No of free items      :", free_items)
    print("Amount to pay         :", int(amount) if amount.is_integer() else amount)
    print("Discount given        :", int(discount) if discount.is_integer() else discount)
    print("Final amount paid     :", int(final_amount) if final_amount.is_integer() else final_amount)

except FileNotFoundError:
    print("Error: File not found.")

except PermissionError:
    print("Error: Permission denied.")

except Exception as e:
    print("An unexpected error occurred:", e)
```
